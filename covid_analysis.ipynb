Performed using Python, Pandas, Matplotlib, and Seaborn.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Everything installed successfully!")
```

```python
import pandas as pd

df = pd.read_csv("covid_19_clean_complete.csv")

print(df.head())
```

```python
# Convert Date column into proper date format
df['Date'] = pd.to_datetime(df['Date'])

# Group data by date
daily_data = df.groupby('Date')[['Confirmed', 'Deaths', 'Recovered']].sum()

# Display first rows
print(daily_data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))

sns.lineplot(x=daily_data.index,
             y=daily_data['Confirmed'])

plt.title("Daily COVID-19 Confirmed Cases")
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(x=daily_data.index,
             y=daily_data['Deaths'],
             color='red')

plt.title("Daily COVID-19 Death Cases")
plt.xlabel("Date")
plt.ylabel("Deaths")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(x=daily_data.index,
             y=daily_data['Recovered'],
             color='green')

plt.title("Daily COVID-19 Recovery Cases")
plt.xlabel("Date")
plt.ylabel("Recovered Cases")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(14,7))

plt.plot(daily_data.index,
         daily_data['Confirmed'],
         label='Confirmed')

plt.plot(daily_data.index,
         daily_data['Deaths'],
         label='Deaths')

plt.plot(daily_data.index,
         daily_data['Recovered'],
         label='Recovered')

plt.title("COVID-19 Overall Trends")

plt.xlabel("Date")
plt.ylabel("Number of Cases")

plt.legend()

plt.xticks(rotation=45)

plt.show()
```

## Conclusion

This project analyzes COVID-19 data using Python libraries such as Pandas, Matplotlib, and Seaborn. The dataset was cleaned and grouped by date to study trends in confirmed cases, deaths, and recoveries. Different visualizations were created to better understand the spread and impact of COVID-19 over time. The analysis shows that confirmed cases increased rapidly during the pandemic, while recovery rates also improved gradually.
